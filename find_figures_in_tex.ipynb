```python
def get_file_by_extension(file_prefix, extension):
    """Get the file names with extension of 'extension'
    """
    process = subprocess.Popen('ls '+file_prefix+'/*.'+extension,
                     shell=True,
                     stdout=subprocess.PIPE, 
                     stderr=subprocess.PIPE,
                     universal_newlines=True)
    stdout, stderr = process.communicate()
    if (stderr != ''):
        raise OSError(stderr)
    files = stdout.split('\n')
    return files
```

```python
def get_image_files(file_prefix):
    """Get images file name"""
    image_extensions = ['pdf', 'eps', 'ps', 'png', 'jpg']
    files = []
    for ext in image_extensions:
        try:
            files.extend(get_file_by_extension(file_prefix, ext))
        except OSError:
            continue
    return files
```

```python
def find_blocks(full_content, block_name):
    """find all blocks with block_name and block_name + * """
    # first split the text
    text_list = full_content.split(r'\begin{')
    found_blocks = []
    _continued = False
    for _text in text_list:
        if (_continued):
            pattern = r"\\end\{\s*"+block_name+"[\s\*]*\}"
            if (re.search(pattern, _text) is not None):
                pattern = r"[\s\S]*\\end\{\s*"+block_name+"[\s\*]*\}"
                _block_con += re.search(pattern, _text).group()
                _continued = False
                found_blocks.append(_block_con)
                continue
            else: 
                _block_con += _text
                continue
        else:
            _block_con = ""
        pattern = block_name
        if(re.match(pattern, _text) is None):
            continue
        pattern = block_name+r"[\s\*]*\}[\s\S]*\\end\{\s*"+block_name+r"[\s\*]*\}"
        _block_con_re = re.search(pattern, _text)
        if (_block_con_re is None):
            _continued = True
            _block_con += _text
        else:
            _block_con += _block_con_re.group()
            found_blocks.append(_block_con)
            
    return found_blocks
```

```python
def find_figure_blocks(full_content):
    """find all figures in the text"""
    return find_blocks(full_content, 'figure')
```

```python
def file_name_without_prefix(file_prefix, files):
    """remove the prefix of files"""
    files_out = []
    for ff in files:
        if (re.search(re.escape(file_prefix), ff) is None):
            print("Warning: no file prefix: "+file_prefix)
        ff_out = re.sub(re.escape(file_prefix), '', ff)
        while (re.match(r'/', ff_out) is not None):
            ff_out = ff_out[1:]
        files_out.append(ff_out)
    return files_out
```

```python
def pair_figures_blocks(fig_files, fig_blocks):
    """make links between figure file names and figure blocks
    fig_files should not contain prefix directories or ./ at beginning
    """
    image_extensions = ['pdf', 'eps', 'ps', 'png', 'jpg'] # should globalize it
    link_dict = {}
    for _block in fig_blocks:
        for _ext in image_extensions:
            pattern = r"[a-zA-Z0-9\w]*\."+ _ext
            _filename_re = re.search(pattern, _block)
            if(_filename_re is not None):
                _filename = _filename_re.group()
                while(re.match(r'./', _filename) is not None):
                    _filename = _filename[2:]
                if (_filename in fig_files):
                    link_dict[_filename] = _block
            else:
                continue
    return link_dict
```

```python
def get_content_in_curly(a_str, description=None):
    """Get the (first) content in the curly braces, with description
    if description is not define, should return the first { content
    """
    _bra = r'{'
    _ket = r'}'
    _bra_count = 0
    _ket_count = 0
    if(description==None):
        pass
    else:
        pattern = description+r"\s*\{"
        _search_re = re.search(pattern, a_str)
        if (_search_re is None):
            print("Warning: No match found!")
            return None
        else:
            _bra_count += 1
            _begin_position = _search_re.end()
            while(_bra_count!=_ket_count):
                _search_ket_re = re.search(_ket, a_str[_begin_position :])
                _begin_position_bk = _begin_position
                if (_search_ket_re is None):
                    if (debug):
                        print(a_str)
                        print(a_str[_begin_position :])
                    raise ValueError("Cannot find the content in the curly braces.")
                else:
                    _ket_count += 1
                    _end_position = _begin_position+_search_ket_re.start()
                if (debug):
                    print(a_str[_begin_position : _end_position])
                _search_bra_re = re.search(_bra, a_str[_begin_position : _end_position])
                while (_search_bra_re is not None):
                    _bra_count += 1
                    _begin_position += _search_bra_re.end()
                    _search_bra_re = re.search(_bra, a_str[_begin_position : _end_position])
                _begin_position = _begin_position_bk + _search_ket_re.end()
                if (debug):
                    print(_bra_count, _ket_count)
                    print(a_str[_begin_position:])
            _begin_position = _search_re.end()
        return a_str[_begin_position : _end_position]
                    
def get_figure_info(fig_link):
    """Get captions and labels frome figure blocks"""
    caption_fig = {}
    label_fig = {}
    for _file in fig_link.keys():
        caption_fig[_file] = get_content_in_curly(fig_link[_file], description='caption')
        label_fig[_file] = get_content_in_curly(fig_link[_file], description='label')
    return caption_fig, label_fig
```
